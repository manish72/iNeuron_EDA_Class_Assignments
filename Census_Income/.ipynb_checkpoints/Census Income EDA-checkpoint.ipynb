### copying original data into another variable

```python
data_copy = data.copy()
```

### Checking null values

```python
data_copy.isnull()
```

### No null or missing values are present in the dataset

```python
data_copy.isnull().sum().sum()
```

### Checking duplicate rows inside dataset

```python
data_copy[data_copy.duplicated()].shape
```

### Total 24 duplicate rows are present inside dataset, so removing those duplicate rows

```python
data_copy = data_copy[~data_copy.duplicated()]
```

### Checking unique values in each column

```python
for i in data_copy.columns:
    print(i)
    print(data_copy[i].unique())
    print("====================================================================")
```

### Seperating numerical and categorical features

```python
numerical_features = [col for col in data_copy.columns if data_copy[col].dtype == 'int64']
```

```python
categorical_features = [col for col in data_copy.columns if data_copy[col].dtype != 'int64']
```

### data with numerical features

```python
data_copy[numerical_features]
```

### data with categorical features

```python
data_copy[categorical_features]
```

### It is been observed that categorical data is having space at the start of the data, which is not readable. So, striping the entire categorical data

```python
for i in categorical_features:
    data_copy[i] = data_copy[i].str.strip()
```

### Filtering the data where workclass,occupation,native-country columns not having the value '?' and removing those data

```python
data_copy = data_copy[(data_copy['workclass']!='?') & (data_copy['occupation']!='?') & (data_copy['native-country']!='?')].reset_index()
```

```python
data_copy = data_copy.drop(columns='index')
```

```python
data_copy[categorical_features].shape
```

### 13. Calculate total male and female people for each race category

```python
data_copy
```

```python
race_sex_count = data_copy.groupby(['race','sex']).count().reset_index()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
race_sex_count
```

### Only for the race 'Amer-Indian-Eskimo'

```python
df = race_sex_count[race_sex_count['race'] == 'Amer-Indian-Eskimo']
fig = px.pie(df, values='workclass',title='Population of European continent',hole=0.5)
fig.show()
```

### For all the race categories

```python
fig = make_subplots(
    rows=1, cols=5,
    specs=[[{"type": "domain"},{"type": "domain"},{"type": "domain"},{"type": "domain"},{"type": "domain"}]],
    subplot_titles=list(race_sex_count['race'].unique()),
    x_title="Figures show percentages of female(0) and male(1) people in different race column"
)

for i,race in enumerate(list(race_sex_count['race'].unique())):
    df = race_sex_count[race_sex_count['race'] == race]
    fig.add_trace(go.Pie(values=df['workclass'],hole=0.3),
              row=1, col=i+1)
fig.update_layout(height=300, legend_title_text = "Sex")
fig.show()
```
