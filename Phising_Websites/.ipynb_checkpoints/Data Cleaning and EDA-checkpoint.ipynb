```python
from scipy.io import arff
import pandas as pd
import numpy as np
import cufflinks as cf
```

```python
cf.go_offline()
```

```python
data = arff.loadarff('Training Dataset.arff')
df = pd.DataFrame(data[0])
```

```python
df.head()
```

### Decoding the entire data to UTF-8 format

```python
for i in df.columns:
    df[i] = df[i].str.decode('utf-8') 
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all').T
```

### checking duplicate values

```python
df[df.duplicated()]
```

### Calculating the sum of null values in the dataset

```python
df.isnull().sum().sum()
```

### Copying the data to different dataframe to preserve original data

```python
df_copy = df.copy()
```

### Removed duplicates

```python
df_copy = df_copy[~df_copy.duplicated()]
```

```python
df_copy
```

```python
df_copy.info()
```

### checking duplicates in copy dataframe

```python
df_copy[df_copy.duplicated()]
```

### Checking NaN values are available or not in entire dataset

```python
df_copy.dropna()
```

### Since Attribute Characteristics are of type Object, converting all the data into integer type (as per problem statement)

```python
for i in df_copy.columns:
    df_copy[i] = df_copy[i].astype('int')
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

### find the phising websites data which is having IP address in the URL and long URLs

```python
df_copy[(df_copy['having_IP_Address'] == -1) & (df_copy['URL_Length'] == -1)].reset_index()
```

### Find phising websites which is having sub-domains, having_At_Symbol and with non-standard port

```python
df_copy[(df_copy['having_Sub_Domain'] == -1) & (df_copy['having_At_Symbol'] == -1) & (df_copy['port'] == -1)]
```

```python
symbol_port = df_copy.groupby('having_At_Symbol')['port'].sum().reset_index()
```

```python
symbol_port
```

### Non-interactive plot

```python
sns.barplot(x='having_At_Symbol', y='port', data=symbol_port)
plt.title("-1 represents phising websites having non-standard port")
```

### Interactive plot

```python
symbol_port.iplot(kind='bar',x='having_At_Symbol',y='port',xTitle='having_At_Symbol',yTitle='Non-Standard-Port',title='-1 in x-axis represents about phising websites')
```
